```python
# 6 Creating a category for calories of high, medium and low & others below 

def calfun (cal, high, low):
    if cal > high:
        return("High")
    if cal > low:
        return("Medium")
    return("Low")

newcol = ffnut4["NUTR_CALORIES_g"].apply(calfun, args=[600, 120])
newcol
```

```python
ffnut4["CATEGORICAL_CAL"] = newcol
```

```python
ffnut4.head()
```

```python
# more transformation - here I am making a grouping variable for the food group column 
# I noted in this variable that foods like meats or veg, etc seem to start with the first word being that food item and then 
# the rest of the text being additional description 
# because of this, I am creating a variable of the first word 


ffnut4["FOOD_CAT1"] = ffnut4['NUTR_FOOD_NAME'].astype(str).apply(lambda x: x.split()[0])
ffnut4["FOOD_CAT1"] = ffnut4["FOOD_CAT1"].str.replace(',','')
ffnut4
```

```python
# More 6 Creating a category for common foods 

def foodfun (food):
    if any(s in food for s in ('Chicken', 'Beef', 'Turkey', 'Lamb', "Veal", "Fish")):
        return("Meat")
    if any(s in food for s in ('Eggplant', 'Califlower', 'Brocoli', 'Artichokes', "Arugula", "Aspargus","Carrots", "Corn")):
        return("Vegetable")
    if any(s in food for s in ('Apples', 'Apricots', 'Grapes', 'Oranges', "Bananas", "Peaches", "Watermelon", "Strawberry")):
        return("Fruit")
    return("Not Categorized")

newcol2 = ffnut4["FOOD_CAT1"].apply(foodfun)
newcol2
```

```python
# adding to the ffnut4 dataframe 
ffnut4["FOOD_CAT2"] = newcol2
```
