```python
# loading libraries 

import numpy as np
import pandas as pd

import statistics 

import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

import requests
```

```python
# first few steps - I am following the text book example from wrangling
# creating a variable for wiki info 

wiki_info = "https://en.wikipedia.org/wiki/Reference_Daily_Intake"
```

```python
# using requests to get the info to the created variable response

response = requests.get(wiki_info)
```

```python
type(response)
```

```python
# using code to determine if the request was a success or not 
def status_ck(r):
        if r.status_code == 200:
            print("Success")
            return 1
        else: 
            print("Failed")
            return -1
        
        
status_ck(response)
```

```python
# checking the encoding

def encoding_ck (r):
    return (r.encoding)
```

```python
encoding_ck(response)
```

```python
# using the encoding and creating a variable contents 
def decode_content (r, encoding):
    return (r.content.decode(encoding))

contents = decode_content(response, encoding_ck(response))
```

```python
# what is contents? 

type(contents)
```

```python
# parsing contents with beautiful soup 

soup = BeautifulSoup(contents, 'html.parser')
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['Nutrient', 'DV'])

# Collecting data
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
      
    if(columns != []):
        Nutrient = columns[0].text.strip()
        DV = columns[1].text.strip()
  

        df = pd.concat([df, pd.DataFrame({'Nutrient': [Nutrient],'DV':[DV]})])
        #df = df.append({'Nutrient': Nutrient,  'DV': DV}, ignore_index=True)
df
```

Ethical considerations:
The Daily Value data is from a public web site and though Wikipedia is very popular, it is not a source for scientific information or monitored by the institutions that create daily nutrition value guidelines. As such, there could be misinformation that I could be inadvertently spreading because, though it would be good to double check the information, I haven’t yet planned to verify the information on wikipedia against a government site. 

Another ethical consideration is that, in general, food guidelines use a single value of 2000 calories as a way to speak to the whole population in terms of daily caloric intake – with an underlying assumption that we are all the same age, size, gender and same need in terms of caloric intake. It is unusual especially in that in many other aspects of our daily lives, we expect people to have a range of values (heart rate, blood pressure, there are many examples). But for some reason, daily calories are typically a single value and moreover, a single target that some may be considering when planning meals or how they approach food in general. However, in reality, people vary quite a bit on what is a healthy amount to eat and vary even more on exercise. An big ethical concern is adding to idea that you can use these nutrition guidelines effectively – when in fact, they themselves could be misleading and misrepresenting what is healthy for the population by not considering the variation of each individual. I think this is an important consideration. Likewise, when doing direct comparisons between foods and their nutritional daily value, I am expecting to do calculations to ensure I am making appropriate adjustments – any results would have to be double checked in order to ensure no error took place. 

Also, health information is constantly changing and, often times, improving. I believe these guidelines are current, but I am not update to date on what is the current message in terms of nutritional values and government recommendations.
